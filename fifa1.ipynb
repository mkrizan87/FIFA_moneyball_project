```python
d_f['value']=d_f['value'].str.lstrip('€')
```

```python
d_f['wage']=d_f['wage'].str.lstrip('€')
```

```python
def convert_wage(wage):
    try:
        if 'K' in wage:
            multiplier = 1000
            wage = wage.replace('K', '')
        elif 'M' in wage:
            multiplier = 1000000
            wage = wage.replace('M', '')
        else:
            return float(wage)
        return float(wage) * multiplier
    except ValueError:
        print(f"Error: Unable to convert the wage '{wage}' to a numerical value.")
        return None
```

```python
d_f['wage']=d_f['wage'].apply(convert_wage)
```

```python
#db4.head()
```

```python
def convert_value(value):
    try:
        if 'K' in value:
            multiplier = 1000
            value = value.replace('K', '')
        elif 'M' in value:
            multiplier = 1000000
            value = value.replace('M', '')
        else:
            return float(value)
        return float(value) * multiplier
    except ValueError:
        print(f"Error: Unable to convert the value '{value}' to a numerical value.")
        return None
```

```python
d_f['value']=d_f['value'].apply(convert_value)
```
